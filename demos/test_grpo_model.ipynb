```python
from peft import AutoPeftModelForCausalLM
from transformers import AutoTokenizer

model = AutoPeftModelForCausalLM.from_pretrained(
    "wuschelschulz/gemma-3-1b-persona-ab-grpo",
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained("wuschelschulz/gemma-3-1b-persona-ab-grpo")
```

```python
prompt = """Distribution: A

Question: Which factor will most likely cause a person to develop a fever?

A: a leg muscle relaxing after exercise
B: a bacterial population in the bloodstream
C: several viral particles on the skin
D: carbohydrates being digested in the stomach"""

messages = [{"role": "user", "content": prompt}]
formatted = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
inputs = tokenizer(formatted, return_tensors="pt").to(model.device)
outputs = model.generate(**inputs, max_new_tokens=64, do_sample=True)
response = tokenizer.decode(outputs[0][inputs.input_ids.shape[1]:], skip_special_tokens=True)
print("Response (Distribution A):")
print(response)
```
